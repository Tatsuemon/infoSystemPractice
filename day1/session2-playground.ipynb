## OPTIONAL 2

Between the iris setosa and the iris versicolor, which has the largest petal surface? (surface $\approx$ width x length)

ヒオウギアヤメとブルーフラッグ、どちらが花びらの面積が大きいですか？ （面積 $\approx$ 幅 x 長さ）

```python
# ヒオウギアヤメ
hiou = iris[iris[:, 0]==0, :]
hiou_width = hiou[3]
hiou_length = hiou[4]
hiou_square = (hiou_width*hiou_length).sum()
hiou_square
```

```python
# ブルーフラッグ
blue = iris[iris[:, 0]==1, :]
blue_width = blue[3]
blue_length = blue[4]
blue_square = (blue_width*blue_length).sum()
blue_square
```

```python
if hiou_square < blue_square:
    print('ブルーフラッグ')
elif hiou_square > blue_square:
    print('ヒオウギアヤメ')
else:
    print('同じです')
```
