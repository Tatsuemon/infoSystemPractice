## TASK 1
Plot the petal length for the iris versicolor and the iris virginica.
- Use the `plot` function to create the plot
- Add the adequate labels

ブルーフラッグ種とバージニカ種の花びらの長さをプロットしてください。
- 関数`plot`を使用してプロットを作成する
- 適切なラベルを追加する


```python
plt.plot(iris[iris[:, 0]==1, 3], color=[1,0,0]) # red
plt.plot(iris[iris[:, 0]==2, 3], color=[0,0,1]) # blue
plt.xlabel("samples")
plt.ylabel("length of petal [cm]")
plt.title("blue flag(red) and virginia(blue)")
```
